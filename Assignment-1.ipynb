Write a Python program to display values of y when 'x' is an independent variable having values [-10, 10, step_size = 0.1] and:
1) y = wx + b; where 'w' and 'b' are constants
2) y = x^2
3) y = 1/ (1 + e^(-x)); where ^ means 'to the power'.
4) y = (e^x - e^(-x)) / (e^x - e^(-x))
5) y = g(f(x)) where u = f(x) = wx + b and y = g(u) = 1/ (1 + e^(-u))
6) y = g(f(x)) where f(x) = wx + b, u = f(x), and g(u) = (e^u - e^(-u)) / (e^u + e^(-u))
7) y1 = g1(f1(x)) where f1(x) = w1x + b1, u1 = f1(x), and g1(u1) = 1/ (1 + e^(-u1))
   y2 = g2(f2(x)) where f2(x) = w2x + b2, u2 = f2(x), and g2(u2) = 1/ (1 + e^(-u2))
   y = g3(f3(v)) where f3(v) = w3y1 + w4y2 + b, w = f3(v), and g3(w) = 1/ (1 + e^(-w))

Importing necessary modules

```python
import matplotlib.pyplot as plt
import numpy as np
```

Taking the values of 'x' as an independent variable and 'w' and 'b' as constants

```python
x = np.arange(-10, 10, 0.1)
w = 5
b = 3

plt.figure(figsize=(16,8))
```

1) Function: y = wx + b

```python
y = w * x + b

plt.plot(x, y, label = ' y = wx + b')
plt.title('Function: y = wx + b')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.legend()
plt.show()
```

2) Function: y = x^2

```python
y = x ** 2

plt.plot(x, y, label = ' y = x^2')
plt.title('Function: y = x^2')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.legend()
plt.show()
```

3) Function: y = 1/ (1 + e^(-x))

```python
y = 1 / (1 + np.exp(-x))

plt.plot(x, y, label = ' y = 1 / (1 + e^-x)')
plt.title('Function: y = 1 / (1 + e^-x)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.legend()
plt.show()
```

4) Function: y = (e^x - e^(-x)) / (e^x - e^(-x))

```python
y = (np.exp(x) - np.exp(-x)) / (np.exp(x) - np.exp(-x))

plt.plot(x, y, label = ' y = (e^x - e^-x) / (e^x + e^-x)')
plt.title('Function: y = (e^x - e^-x) / (e^x + e^-x)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.legend()
plt.show()
```

5) Function: y = g(f(x)) where u = f(x) = wx + b and y = g(u) = 1/ (1 + e^(-u))

```python
u = w * x + b
y = 1 / (1 + np.exp(-u))

plt.plot(x, y, label = ' y = g(f(x))')
plt.title('Function: y = g(f(x))')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.legend()
plt.show()
```

6) Function: y = g(f(x)) where f(x) = wx + b, u = f(x), and g(u) = (e^u - e^(-u)) / (e^u + e^(-u))

```python
u = w * x + b
y = (np.exp(u) - np.exp(-u)) / (np.exp(u) + np.exp(-u))

plt.plot(x, y, label = 'y = g(f(x))')
plt.title('Function: y = g(f(x))')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.legend()
plt.show()
```

7) y1 = g1(f1(x)) where f1(x) = w1x + b1, u1 = f1(x), and g1(u1) = 1/ (1 + e^(-u1))<br>y2 = g2(f2(x)) where f2(x) = w2x + b2, u2 = f2(x), and g2(u2) = 1/ (1 + e^(-u2))<br>Function: y = g3(f3(v)) where f3(v) = w3y1 + w4y2 + b, w = f3(v), and g3(w) = 1/ (1 + e^(-w))

```python
w1, b1 = 4, 10
u1 = w1 * x + b1
y1 = 1 / (1 + np.exp(-u1))

w2, b2 = 6, 3 
u2 = w2 * x + b2
y2 = 1 / (1 + np.exp(-u2))

w3, w4 = 2, 3
w = w3 * y1 + w4 * y2 + b
y = 1 / (1 + np.exp(-w))

plt.plot(x, y, label = 'y = g3(f3(v))')
plt.title('Function: y = g3(f3(v))')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.legend()
plt.show()
```
